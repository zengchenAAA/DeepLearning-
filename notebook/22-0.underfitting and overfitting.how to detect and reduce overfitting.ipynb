## Regularization

$ J(\theta) = -\frac{1}{m}\sum_{i=1}^{m}[y_iln\hat y_i + (1-y_i)ln(1-\hat y_i)] + \frac{1}{2}\lambda \sum_{i=1}^{n}||\theta_i||_2)$ 

在上面的式子中 前面的是cross entropy的结果相加  后面是对于网络参数取二范数(其中任何范数都可以)然后相加乘以类似LR的$\lambda 参数$  
式子上不仅minimize loss的值还minimize参数的范数的值，enforce weights close to 0 ,使得曲线比较平滑。这样就可以减少模型的复杂度
