## Cross Entropy

$H(p,q) = \sum p(x)log q(x) $  
$H(p,q) = H(p) + D_{KL}(p|q) $  
$ D_{KL}$ is KL Divergence， the $D_{KL}(p|q) \approx 0 $ when the distribution get closer  
  
the value of the KL Divergence opposed to the value of Entropy.

* P = Q
  * cross entropy = Entropy  
  $ H(p,q) = H(p)$
* for one-hot encoding
  * entropy = 1log1 = 0

## Binary Classification

$ H(P,Q) = -P(cat) log Q(cat) - (1- P(cat))log(1-Q(cat))$
$ P(dog) = (1 - P(cat))$
$$H(P,Q) = - \sum_{i=(cat,dog)} P(i)logQ(i)
       = - P(cat)logQ(cat)-P(dog)logQ(dog)
       = -(ylog(p)+(1-y)log(1-p))
$$  
这里的y表示的是cat的实际的值 p表示网络输出值

```python
import torch
import torch.nn.functional as F 
```

```python
x = torch.randn(1,784)
w = torch.randn(10,784)
```

```python
logits = x@w.t()
```

```python
logits.size()
```

```python
F.cross_entropy(logits, torch.tensor([9])) #第二个参数代表你要查看的第几个下标的crossEntory的值(0-9)
```

```python
pred = F.softmax(logits,dim=1)
```

```python
pred_log = torch.log(pred)
```

```python
F.nll_loss(pred_log,torch.tensor([9]))
```

so CE = softmax + log + nll_loss
